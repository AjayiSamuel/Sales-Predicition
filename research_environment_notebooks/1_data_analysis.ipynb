```python
# to handle datasets
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
%matplotlib inline

# to display all the columns of the df
pd.pandas.set_option('display.max_columns', None)
```

```python
# load dataset
data = pd.read_csv(r'./../dataset/1-raw/assessment_test_data.csv')

# rows and columns of the data
print(data.shape)

# visualise the dataset
data.head()
```

```python
# histogran to evaluate target distribution
data['units_sold'].hist(bins=50, density=True)
plt.ylabel('Frequency')
plt.xlabel('Unit Sold')
plt.show()
```

#### In the notebook we will analyse the dataset to identify the folowing:

- Missing values
- Numerical variables
- Distribution of the numerical variables
- Outliers
- Categorical variables

```python
# Missing Values

# make a list of the variables that contain missing values
vars_with_na = [var for var in data.columns if data[var].isnull().sum()>1]


# print the variable name and the percentage of missing values
for var in vars_with_na:
    print(var, np.round(data[var].isnull().mean(),2)*100, '% missing values')
    
print(f"We can observe {len(vars_with_na)}/{data.shape[1]} features contain missing values")
```

```python
def analyse_na_value(df, var):
    df = df.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    df[var] = np.where(df[var].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    df.groupby(var)['units_sold'].median().plot.bar()
    plt.title(var)
    plt.show()
    
for var in vars_with_na:
    analyse_na_value(data, var)
```

```python
# list of numerical variables
num_vars = [var for var in data.columns if data[var].dtypes != 'O']

print('Number of numerical variables: ', len(num_vars))

# visualise the numerical variables
data[num_vars].head()
```

Here, we observe that all the features apart form the *data* feature are numerical this will limit the feature engineering (Data preprocessing) required.

```python
# List of categorical variables

categorical = [var for var in data.columns if data[var].dtype=='O']

print('There are {} categorical variables'.format(len(categorical)))
```

```python
data[categorical].head()
```

The Categorical variable is a unique one, as it is a date variable. we have special ways to edit date variables to extract added value.

```python
import datetime

data['sales_date'] = pd.to_datetime(data['date'])
data[['date', 'sales_date']].head()
```

```python
data['sales_date'].iloc[0], data['sales_date'].iloc[data.shape[0]-1]
```

```python
print("Dataset duration contains sales, over a duration of", data['sales_date'].iloc[data.shape[0]-1] - data['sales_date'].iloc[0] )
```

We observe our dataset contains sales information over the course of 2 years and 23 days.

```python
# Extracting values form dates in different formats

# day - numeric from 1-31
data['dt_day'] = data['sales_date'].dt.day
# Extracting week of year from date, varies from 1 to 52
data['dt_week'] = data['sales_date'].dt.week
# Extracting month from date - 1 to 12
data['dt_month'] = data['sales_date'].dt.month
# Extracting quater from date - 1 to 4
data['dt_quater'] = data['sales_date'].dt.quarter
# Extracting day name from Sunday - Saturday
data['dt_dayofweek'] = data['sales_date'].dt.day_name()
# Extracting is weekend from date from date - 1 or 2 (Whether the transaction was done on a weekday or weekend)
data['dt_is_weekend'] = np.where(data['dt_dayofweek'].isin(['Sunday', 'Saturday']), 1,0)
```

```python
data[['dt_day', 'dt_week','dt_month', 'dt_quater', 'dt_is_weekend', 'dt_dayofweek']].head()
```

It is important to note that values extracted from dates will be treated a categorical variables as they are nominal data lacking in any form of order or rank.

```python
# Discrete variables 
# Let us identify numerical variables/features that have finite number of values, this means they are not continous variables

#  list of discrete variables
discrete_vars = [var for var in num_vars if len(data[var].unique())<20]

print('Number of discrete variables: ', len(discrete_vars))
```

```python
# let's visualise the discrete variables
data[discrete_vars].head()
```

```python
def analyse_discrete(df, var):
    df = df.copy()
    df.groupby(var)['units_sold'].median().plot.bar()
    plt.title(var)
    plt.ylabel('units_sold')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(data, var)
```

```python
# Continous Variables

# list of continuous variables
cont_vars = [var for var in num_vars if var not in discrete_vars]

print('Number of continuous variables: ', len(cont_vars))
```

```python
# let's visualise the continuous variables
data[cont_vars].head()
```

```python
# Let's go ahead and analyse the distributions of these variables
def analyse_continous(df, var):
    df = df.copy()
    df[var].hist(bins=20)
    plt.ylabel('Frequency')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_continous(data, var)
```

Interesting we observe days with Zero units of items sold have the highest frequency

```python
# Visualising the continuous variables
data[cont_vars].head()
```

```python
# Exploring the relationship between Units Sold and the transformed variables
# with more detail
def transform_analyse_continous(df, var):
    df = df.copy()
    
    # log does not take negative values so skipping those variables
    if 0 in data[var].unique():
        pass
    else:
        # log transform
        df[var] = np.log(df[var])
        df['units_sold'] = np.log(df['units_sold'])
        plt.scatter(df[var], df['units_sold'])
        plt.ylabel('units_sold')
        plt.xlabel(var)
        plt.show()
    
for var in cont_vars:
    if var !='units_sold':
        transform_analyse_continous(data, var)
```

```python
# Outliers

# Boxplots to visualise outliers in the continuous variables 

def find_outliers(df, var):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()
    
for var in cont_vars:
    find_outliers(data, var)
```

A decent amount of outliers can be observed above.

```python
cat_vars = [var for var in data.columns if var not in num_vars+['date', 'sales_date']]

print('Number of categorical variables: ', len(cat_vars))
```

```python
# Values of the categorical variables
data[cat_vars].head()
```

```python
# Cardinality of Categorical variables

for var in cat_vars:
    print(var, len(data[var].unique()), ' categories')
```

```python
# Rare labels

# Finding the rare labels in the daily transactions
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()
    tmp = df.groupby(var)['units_sold'].count() / len(df)
    return tmp[tmp<rare_perc]

for var in cat_vars:
    print(analyse_rare_labels(data, var, 0.01))
    print()
```

```python
# Relationship between our categorical variables and Unit sold per day

for var in cat_vars:
    analyse_discrete(data, var)
```
